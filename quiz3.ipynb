1.리스트 내포 - 피타고라스 정리

```python
print([(x,y,z) for x in range(1,101,1) for y in range(x,101,1) for z in range(1,1001,1) if z**2 == x**2+y**2])
```

2.리스트 내포 - 소수 만들기

```python
print(list(filter(None, [tmp[-1] if all([False if tmp[-1] != z and tmp[-1] % z == 0 else tmp[-1] for z in tmp]) else None
                            for tmp in [[j for j in range(2,i+1)] for i in range(2,1001)]])))
```
